# Step 1: Data Generation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import warnings
warnings.filterwarnings("ignore")
```

# Step 2: Data Preparation

```python
np.random.seed(5)
hours_studied=np.random.uniform(0,10,100)
exam_scores=50 + 5 *hours_studied + np.random.normal(0,2,100)
```

```python
data=pd.DataFrame({'Hours_Studied': hours_studied,'Exam_Scores':exam_scores})

data.head()                  
```

```python
data.info()
```

# Step 3: Data Splitting

```python
X = data['Hours_Studied']
y = data['Exam_Scores']
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.2,random_state=75)
```

# Step 4: Linear Regression Modeling

```python
import pandas as pd
from sklearn.linear_model import LinearRegression


# Assuming X_train and y_train are Series
X_train = X_train.to_frame().reset_index(drop=True)  # Convert Series to DataFrame and reset index
y_train = y_train.to_frame().reset_index(drop=True)  # Convert Series to DataFrame and reset index if y_train is also a Series
# Example of fitting a model
model = LinearRegression()
model.fit(X_train, y_train)
```

# Step 5: Model Evaluation

```python
X_test = X_test.to_frame().reset_index(drop=True)  # Convert Series to DataFrame and reset index
y_test = y_test.to_frame().reset_index(drop=True)  
```

```python
# make predictions on testing data

y_pred=model.predict(X_test)


                     
```

```python
# Evaluet the model's performance

mse=mean_squared_error(y_test, y_pred)
r2=r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```

# Step 6: Visualization and Interpretation

```python
# plot the regression line and data point
plt.scatter(X_test,y_test,color='blue',label='Actual Data')
plt.plot(X_test,y_pred,color='red',label='Regression Line',linewidth=2)
plt.xlabel('Hours Stidied')
plt.ylabel('Exam Score')
plt.legend()
plt.title('Simple Linear Regression: Hours Studied vs. Exam Scores')
plt.show()
```

```python
# interpret the model's coefficients

slope= model.coef_[0]
intercept = model.intercept_


print(f'Slope (Coefficient): {slope}')
print(f'Intercept: {intercept}')
```
